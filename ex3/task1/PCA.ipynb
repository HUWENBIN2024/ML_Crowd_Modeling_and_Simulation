```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data_points = np.loadtxt("../data/pca_dataset.txt", dtype=float)

data_point_mean = np.mean(data_points, axis=0)
centered_data_points = np.subtract(data_points, data_point_mean)
centered_data_points
```

```python
u, s, v = np.linalg.svd(centered_data_points)

uu = u[:, 0:1] # (100, 1)
vv = v[:, 0:1] # (2, 1)
ss = np.zeros((u.shape[0], v.shape[0]))
ss[:len(s), :len(s)] = np.diag(s)
ss = ss[0:1, 0:1] # (1, 1)

# xx = uu @ ss @ vv.T
# xx += np.mean(data_points, axis=0)
```

```python
energy = np.square(s) / np.sum(np.square(s))

# Obtain direction of energy vector
direction = v.T.dot(energy)
print(direction)

plt.scatter(x=data_points[:,0:1], y=data_points[:, 1:2])

for i in range(len(s)):
    plt.arrow(0, 0, energy[i] * v[i, 0], energy[i] * v[i, 1], width=0.04, color='red')
```
