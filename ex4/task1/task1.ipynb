## Task 1: Vector fields, orbits, and visualization

```python
# define notebook parameters
w = 6
Y, X = np.mgrid[-w:w:100j, -w:w:100j]

```

```python
def plot(matrix):
    # example for Euler's method to construct and plot a trajectory over the stream plot
    y0 = np.array([-1, 1.5])
    time = np.linspace(0, 50, 1000)
    yt, time = solve_euler(lambda y: matrix@y, y0, time)
    
    # example linear vector field A*x
    ax0 = plot_phase_portrait_linear(matrix, X, Y)
    # then plot the trajectory over it
    ax0.plot(yt[:, 0], yt[:, 1], c='red', label='Trajectory at y0=(-1,1.5)')
    ax0.set_xlim(-6, 6)
    ax0.set_ylim(-6, 6)
    # prettify
    ax0.legend()
    ax0.set_aspect(1)
    plt.show()
```

### Stable focus:

```python
# example matrix
alpha = 2
A = np.array([[alpha,-alpha*4],
              [4, -6]])
plot(A)
```

### Stable node:

```python
# example matrix
alpha = 0.2
A = np.array([[-alpha,0],
              [0, -1]])
plot(A)
```
