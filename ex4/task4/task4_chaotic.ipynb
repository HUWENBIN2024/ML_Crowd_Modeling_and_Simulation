### Part 2 : Lorenz attractor
Given:  a Lorenz system: 
</br> $\frac{\mathrm{dx}}{\mathrm{d}t} = \sigma (y - x)$  &nbsp;  $\frac{\mathrm{dy}}{\mathrm{d}t} = x (\rho - z) - y $ &nbsp;   $\frac{\mathrm{dz}}{\mathrm{d}t} = xy - \beta z$</br>
To Show: </br>
a trajectory of the Lorenz attractor starting at (10,10,10)

part 2.1</br>
$\sigma = 10, \beta = 8/3, \rho = 28$

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6), sharey=True,subplot_kw={'projection': '3d'})
x1=plot_lorenz_trajectory([10, 10, 10],10,8/3, 28,ax=ax1)
x2=plot_lorenz_trajectory([10+1e-8, 10, 10],10,8/3, 28,ax=ax2)
```

```python
comparsion(x1, x2)
```

part 2.2</br>
$sigma = 10, \beta = 8/3, \rho = 0.5$

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6), sharey=True,subplot_kw={'projection': '3d'})
x1=plot_lorenz_trajectory([10, 10, 10],10,8/3, 0.5,ax=ax1)
x2=plot_lorenz_trajectory([10+1e-8, 10, 10],10,8/3, 0.5,ax=ax2)
```

```python
comparsion(x1, x2)
```
