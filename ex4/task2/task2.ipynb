```python
import numpy as np
import matplotlib.pyplot as plt
```

Create the bifurcation diagram for system: 
$$\dot{x} = \alpha - x^2$$

```python
alpha = np.linspace(0, 1, 100)
plt.xlim(-1, 1)
plt.ylim(-1, 1)
plt.xlabel(r"$\alpha$")
plt.ylabel(r"$x$")
plt.plot(alpha,  np.sqrt(alpha), color="orange", label="Stable")
plt.plot(alpha, -np.sqrt(alpha), color="blue", label="Unstable")
plt.title(r"$\dot{x} = \alpha - x^2$")
plt.grid(True)
plt.legend()
plt.show()
```

Create the bifurcation diagram for system: 
$$\dot{x} = (\alpha - 3) - 2x^2$$

```python
alpha = np.linspace(3, 5, 100)
plt.xlim(1, 5)
plt.ylim(-1, 1)
plt.xlabel(r"$\alpha$")
plt.ylabel(r"$x$")
plt.plot(alpha,  np.sqrt((alpha - 3) / 2), color="orange", label="Stable")
plt.plot(alpha, -np.sqrt((alpha - 3) / 2), color="blue", label="Unstable")
plt.title(r"$\dot{x} = (\alpha - 3) - 2x^2$")
plt.grid(True)
plt.legend()
plt.show()
```
